```python
import pandas as pd  
df = pd.read_csv('titanic.csv')
```

univariate analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.hist(df['age'], bins=10)
plt.xlabel('age')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
#Bar plot of who
sns.countplot(x='who', data=df)
plt.xlabel('who')
plt.ylabel('Count')
plt.title('Bar Plot')
plt.show()
```

bi-variate analysis

```python
plt.scatter(df['pclass'], df['age'])
plt.xlabel('pclass')
plt.ylabel('age')
plt.title('Scatter Plot')
plt.show()
```

```python
sns.boxplot(x='class', y='fare', data=df)
plt.xlabel('class')
plt.ylabel('fare')
plt.title('Box Plot')
plt.show()
```

check for categorical columns and perform

```python
encoded_df=pd.get_dummies(df,columns=['class'])
```

splitting data into dependent and independent variables

```python
X_numerical = df.select_dtypes(include=['float64', 'int64'])
y = encoded_df['survived']
```

scale the independent variables

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X_numerical)
```

split the data into training and testing

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```
